```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn import linear_model
plt.style.use('_classic_test_patch')
```

```python
canada_df = pd.read_csv('./canada_per_capita_income.csv')
canada_df
```

```python
plt.scatter(canada_df['year'],canada_df['per capita income (US$)'])
plt.title('Canada_per_capita_income')
plt.xlabel('Year')
plt.ylabel('per capita income (US$')
plt.legend()
plt.tight_layout()


plt.show()
```

```python
regression=linear_model.LinearRegression()
regression.fit(canada_df[['year']],canada_df['per capita income (US$)'])
```

```python
regression.predict([[2020]])
```

```python
plt.scatter(canada_df['year'],canada_df['per capita income (US$)'])
plt.plot(canada_df['year'],regression.predict(canada_df[['year']]))
plt.title('Canada_per_capita_income')
plt.xlabel('Year')
plt.ylabel('per capita income (US$')
plt.legend()
plt.tight_layout()


plt.show()
```
